```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

url = 'https://raw.githubusercontent.com/jayparab26/Covid-19-Data-Analysis/main/covid_19_india.csv'
df = pd.read_csv(url)
```

# **Uni-Variate Analysis**

```python
df.describe()
```

```python
filtered_df=df[df['Confirmed']>0]
sns.histplot(filtered_df['Confirmed'], log_scale=True, kde=True)
```

# **Bi-Variate Analysis**

```python
sns.scatterplot(x='Confirmed', y='Cured', data=df)
```

```python
sns.scatterplot(x='Confirmed', y='Deaths', data=df)
```

# **Multi-Variate Analysis**

```python
sns.scatterplot(data=df, x='Confirmed', y='Deaths', hue='Cured')
```

```python
# 1. Calculate the correlation values for your numerical columns
matrix_data = df[['Confirmed', 'Deaths', 'Cured']].corr()

# 2. Pass this clean data into the heatmap
sns.heatmap(matrix_data, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1)
```
